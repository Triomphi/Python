## Concatenation

Concatenation basically glues together DataFrames. Keep in mind that dimensions should match along the axis you are concatenating on. You can use **pd.concat** and pass in a list of DataFrames to concatenate together:

```python
pd.concat([df1, df2, df3])
```

```python
pd.concat([df1,df2,df3],axis=1)
```

## Merging

The **merge** function allows you to merge DataFrames together using a similar logic as merging SQL Tables together. For example:

```python
pd.merge(data_1, data_2)
```

Or to show a more complex example:

```python
new_data_a = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
new_data_b = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
new_data_a
```

```python
new_data_b
```

```python
pd.merge(new_data_a, new_data_b)
```

```python
pd.merge(new_data_a, new_data_b, how="outer")
```

```python
pd.merge(new_data_a, new_data_b, how='outer', on=['key1', 'key2'])
```

```python
pd.merge(new_data_a, new_data_b, how='right', on=['key1', 'key2'])

#If the merge is to the right, the columns to the right isn't affected in the merge. 
#The columns to the left however are affected
```

```python
pd.merge(new_data_a, new_data_b, how='left', on=['key1', 'key2'])
#If the merge is to the left, the columns to the left isn't affected in the merge. 
#The columns to the right however are affected
```

## Joining
Joining is a convenient method for combining the columns of two potentially differently-indexed DataFrames into a single result DataFrame.

```python
data_1 = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

data_2 = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```python
data_1.join(data_2, how="left")
```

```python
data_1.join(data_2, how="inner")
```

```python
data_1.join(data_2, how="right")
```

```python
data_1.join(data_2, how='outer')
```
